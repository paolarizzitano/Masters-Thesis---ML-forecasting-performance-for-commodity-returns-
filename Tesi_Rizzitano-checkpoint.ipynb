## Import and visualize commodity prices 

```python
# define commodity tickers

tickers = {
    "Gold": "GC=F",
    "Natural Gas": "NG=F",
    "Corn": "XC=F",
    }

# download monthly data for the last 10 years (Jan 1st, 2014 - Oct 1st, 2025)

def download_data(ticker_dict, start="2014-12-01", end="2025-10-01"):
    data = {}
    for name, ticker in ticker_dict.items():
        df = yf.Ticker(ticker).history(start=start, end=end)

        # Resample to month-end
        df.index = pd.to_datetime(df.index).tz_localize(None)
        df = df.resample("ME").last()

        data[name] = df[["Close"]].rename(columns={"Close": name})

    return data

raw_data = download_data(tickers)

# combine in a unique dataframe
    
combined = pd.concat(raw_data.values(), axis=1).dropna()
combined.head()
combined = combined.dropna()
combined.index = combined.index.date

# plot each commodity time series of prices

for col in combined.columns:
    plt.figure(figsize=(12,5))
    plt.plot(combined.index, combined[col], color='black')
    plt.title(f"{col}  Time Series of Prices")
    plt.xlabel("Date")
    plt.ylabel("Price in USD")
    plt.grid(True)
    plt.show()
```

```python
# plot combined time series 

normalized = combined / combined.iloc[0]  # relative performance wrt price at t0

plt.figure(figsize=(14,7))
for col in normalized.columns:
    plt.plot(normalized.index, normalized[col], label=col)

plt.title("Normalized Commodity Prices (Base = 1)")
plt.xlabel("Date")
plt.legend()
plt.grid(True)
plt.show()
```

## AR (autoregressive) Models 

```python
#gold
# AR(p) setup

y_gold = log_returns["Gold"]

lags = 5
X_gold_AR = pd.DataFrame(index=y_gold.index)

for lag in range(1, lags + 1):
    X_gold_AR[f"Gold_lag{lag}"] = y_gold.shift(lag)

# Drop NaNs (due to lags)
X_gold_AR = X_gold_AR.dropna()
y_gold_AR = y_gold.loc[X_gold_AR.index]

# Expanding window split
n = len(X_gold_AR)
split = int(0.9 * n)

y_gold_forecast_AR = []

for i in range(split, n):

    # Expanding window: 0  i-1
    X_train = X_gold_AR.iloc[:i]
    y_train = y_gold_AR.iloc[:i]

    # Observation to forecast (t)
    X_test = X_gold_AR.iloc[i:i+1]

    # Standardize using only training data
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled  = scaler.transform(X_test)

    # AR estimation via OLS
    model_AR = LinearRegression()
    model_AR.fit(X_train_scaled, y_train)

    # One-step-ahead forecast
    y_hat = model_AR.predict(X_test_scaled)[0]
    y_gold_forecast_AR.append(y_hat)

# Convert to Series with proper index
forecast_index = X_gold_AR.index[split:]
y_gold_forecast_AR = pd.Series(y_gold_forecast_AR, index=forecast_index)

# Evaluation AR(5) - Gold
# ============================

# True out-of-sample returns
y_true = y_gold_AR.iloc[split:]  # r_{t} for t = split ... n-1

# Forecasts from AR model
y_forecast_AR = y_gold_forecast_AR  # gi allineata con y_true

# Random Walk without drift (r_{t} = r_{t-1})
rw_pred = y_gold_AR.shift(1).iloc[split:]
rw_pred.index = y_true.index  # allinea gli indici

# Random Walk with drift (expanding mean)
rw_drift_pred = []
for i in range(split, len(y_gold_AR)):
    rw_drift_pred.append(y_gold_AR.iloc[:i].mean())
rw_drift_pred = pd.Series(rw_drift_pred, index=y_true.index)

# MSPE calculations
MSPE_AR = ((y_true - y_forecast_AR) ** 2).mean()
MSPE_RW = ((y_true - rw_pred) ** 2).mean()
MSPE_RW_drift = ((y_true - rw_drift_pred) ** 2).mean()

# R out-of-sample (Campbell & Thompson)
R2_OoS_vs_RW = 100 * (1 - MSPE_AR / MSPE_RW)           # vs RW no drift
R2_OoS_vs_RW_drift = 100 * (1 - MSPE_AR / MSPE_RW_drift)  # vs RW with drift

# Print results
print("MSPE AR gold:", MSPE_AR)
print("MSPE RW no drift:", MSPE_RW)
print("MSPE RW with drift:", MSPE_RW_drift)
print("R_OoS AR vs RW no drift (%):", R2_OoS_vs_RW)
print("R_OoS AR vs RW with drift (%):", R2_OoS_vs_RW_drift)
```

```python
# plot gold AR(5)
plt.figure(figsize=(14,6))

# True returns
plt.plot(y_true.index, y_true, label="True Gold returns", color="grey")

# AR(5) forecast
plt.plot(y_forecast_AR.index, y_forecast_AR, label="AR(5) forecast", color="red")

# Random Walk without drift
plt.plot(rw_pred.index, rw_pred, label="Random Walk (no drift)", color="blue", alpha=0.6)

# Random Walk with drift
plt.plot(rw_drift_pred.index, rw_drift_pred, label="Random Walk (with drift)", color="green", alpha=0.6)

plt.title("Gold Returns: AR(5) vs Random Walk")
plt.xlabel("Date")
plt.ylabel("Log-Return")
plt.legend()
plt.grid(True)
plt.show()
```

## Insert technical indicators

```python
assets = ["Gold", "Corn", "Natural Gas"]

ma_windows = [1, 3, 6, 9, 12]
momentum_windows = [1, 3, 6, 9, 12]
oslt_k = [1, 3, 6, 9, 12]
oslt_eta = [5, 10]
sr_k = [1, 3, 6, 9, 12]
sr_eta = [1, 2, 3, 4, 5]

lags = [1, 3, 6, 9, 12]
delta = 0.01  # Filtering Rule (1%)

features = []

# --------------------------------
# 1) Momentum
# --------------------------------
for asset in assets:
    price = combined[asset]
    for k in momentum_windows:
        s = (price >= price.shift(k)).astype(np.int8)
        s.name = f"{asset}_MOM_{k}"
        features.append(s)

# --------------------------------
# 2) Moving Average (MA)
# --------------------------------
for asset in assets:
    price = combined[asset]
    for k in ma_windows:
        ma_k = price.rolling(k).mean()
        s = (price >= ma_k).astype(np.int8)
        s.name = f"{asset}_MA_{k}"
        features.append(s)

# --------------------------------
# 3) Oscillator / OSLT (RSI-type)
# --------------------------------
for asset in assets:
    price = combined[asset]
    returns = price.diff()

    for k in oslt_k:
        U = returns.clip(lower=0).rolling(k).sum()
        D = (-returns.clip(upper=0)).rolling(k).sum()
        RSI = 100 * U / (U + D)

        for eta in oslt_eta:
            buy = (RSI <= (50 - eta)).astype(np.int8)
            sell = (RSI >= (50 + eta)).astype(np.int8)

            buy.name = f"{asset}_OSLT_buy_k{k}_eta{eta}"
            sell.name = f"{asset}_OSLT_sell_k{k}_eta{eta}"

            features.append(buy)
            features.append(sell)

# --------------------------------
# 4) SupportResistance (SR)
# --------------------------------
for asset in assets:
    price = combined[asset]

    for k in sr_k:
        roll_max = price.shift(1).rolling(k).max()
        roll_min = price.shift(1).rolling(k).min()

        for eta in sr_eta:
            buy = (price >= (1 + eta/100) * roll_max).astype(np.int8)
            sell = (price <= (1 - eta/100) * roll_min).astype(np.int8)

            buy.name = f"{asset}_SR_buy_k{k}_eta{eta}"
            sell.name = f"{asset}_SR_sell_k{k}_eta{eta}"

            features.append(buy)
            features.append(sell)

# --------------------------------
# 5) Filtering Rule (Alexander, 1961)
# --------------------------------
for asset in assets:
    price = combined[asset]

    for k in lags:
        roll_low = price.rolling(k).min()
        roll_high = price.rolling(k).max()

        buy = (price > (1 + delta) * roll_low).astype(np.int8)
        sell = (price < (1 - delta) * roll_high).astype(np.int8)

        buy.name = f"{asset}_FR_buy_{k}"
        sell.name = f"{asset}_FR_sell_{k}"

        features.append(buy)
        features.append(sell)

# --------------------------------
# DataFrame
# --------------------------------
technical_indicators = pd.concat(features, axis=1)
technical_indicators = technical_indicators.dropna()

print("Technical indicators shape:", technical_indicators.shape)
technical_indicators.to_pickle("technical_indicators.pkl")
```

```python
# Build regression datasets for each asset

assets = ["Gold", "Corn", "Natural Gas"]

commodity_datasets = {} 

for asset in assets:

    # Select technical indicators for this commodity
    asset_indicators = [
        col for col in technical_indicators.columns 
        if col.startswith(asset + "_")
    ]
    
    tech_df = technical_indicators[asset_indicators]

    # Target = next-month return
    ret_df = log_returns[[asset]].shift(-1).rename(columns={asset: "Return"})

    df = ret_df.merge(tech_df, left_index=True, right_index=True, how="inner")
    df = df.dropna()

    commodity_datasets[asset] = df

# Access the datasets

gold_dataset = commodity_datasets["Gold"]
corn_dataset = commodity_datasets["Corn"]
ng_dataset = commodity_datasets["Natural Gas"]

print("Gold dataset:", gold_dataset.shape)
print("Corn dataset:", corn_dataset.shape)
print("Natural Gas dataset:", ng_dataset.shape)

gold_dataset.head()
```

```python
# run univariate OLS for each indicator

def run_ols_forecasts(df):
    """
    df: dataset containing technical indicators + 'Return'
    Returns:
        forecasts_df: dataframe with one forecast column per indicator
    """
    
    indicators = [c for c in df.columns if c != "Return"]
    forecasts = pd.DataFrame(index=df.index)
    
    for col in indicators:
        X = df[[col]]
        X = sm.add_constant(X)
        y = df["Return"]

        model = sm.OLS(y, X).fit()

        # In-sample fitted values as forecasts
        forecasts[f"pred_ret_{col}"] = model.fittedvalues
        
    return forecasts

gold_preds = run_ols_forecasts(commodity_datasets["Gold"])
corn_preds = run_ols_forecasts(commodity_datasets["Corn"])
ng_preds = run_ols_forecasts(commodity_datasets["Natural Gas"])

print("Gold predictions:", gold_preds.shape)
print("Corn predictions:", corn_preds.shape)
print("NatGas predictions:", ng_preds.shape)
```

```python
# average predictions across indicators 

def average_by_rule(forecast_df):
    """
    forecast_df: dataframe with predictions for each indicator
    Returns: dataframe with 5 averaged forecast series (MOM, MA, FR, OSLT, SR)
    """

    rule_groups = {
        "MOM":  [col for col in forecast_df.columns if "_MOM_" in col],
        "MA":   [col for col in forecast_df.columns if "_MA_" in col],
        "FR":   [col for col in forecast_df.columns if "_FR_" in col],
        "OSLT": [col for col in forecast_df.columns if "_OSLT_" in col],
        "SR":   [col for col in forecast_df.columns if "_SR_" in col],
    }

    averaged = pd.DataFrame(index=forecast_df.index)

    for rule, cols in rule_groups.items():
        if len(cols) > 0:
            averaged[f"avg_{rule}"] = forecast_df[cols].mean(axis=1)
        else:
            print(f" Warning: no columns found for rule {rule}")

    return averaged

# synthetic signals == retuns t+1 = f(technicals t) 

gold_avg = average_by_rule(gold_preds)
corn_avg = average_by_rule(corn_preds)
ng_avg = average_by_rule(ng_preds)

gold_avg.head()
```

## OLS Rolling Regression - Gold

```python
# rolling regression setup

n = len(X_gold)
split = int(0.9 * n)  # 90% train 

y_gold_forecast = []

for i in range(split, n):
    # expanding window: 0  i-1
    X_gold_train = X_gold.iloc[:i]
    y_gold_train = y_gold.iloc[:i]
    
    # predict
    X_gold_test = X_gold.iloc[i:i+1]
    
    # regression
    X_gold_train_const = sm.add_constant(X_gold_train)
    X_gold_test_const = sm.add_constant(X_gold_test, has_constant='add')
    
    model = sm.OLS(y_gold_train, X_gold_train_const).fit()
    
    # out-of-sample prediction
    y_gold_hat = model.predict(X_gold_test_const).iloc[0]
    y_gold_forecast.append(y_gold_hat)

# convert to series 

y_gold_forecast = pd.Series(y_gold_forecast, index=y_gold.iloc[split:].index)
```

```python
# True returns
# ============================

y_gold_true = y_gold.iloc[split:]

# ============================
# Random Walk benchmark - no drift
# ============================

rw_pred_gold = y_gold.shift(1).iloc[split:]
rw_pred_gold.index = y_gold_true.index

# Random Walk benchmark - with drift (expanding mean)
# ============================
rw_drift_pred_gold = []

for i in range(split, len(y_gold)):
    # expanding mean up to t-1
    rw_drift_pred_gold.append(y_gold.iloc[:i].mean())

rw_drift_pred_gold = pd.Series(rw_drift_pred_gold, index=y_gold_true.index)

# ============================
# MSPE and R out-of-sample
# ============================

MSPE_model_gold = ((y_gold_true - y_gold_forecast) ** 2).mean()
MSPE_RW_gold    = ((y_gold_true - rw_pred_gold) ** 2).mean()
MSPE_RW_drift_gold = ((y_gold_true - rw_drift_pred_gold) ** 2).mean()

R2_OoS_model_vs_RW_gold = 100 * (1 - MSPE_model_gold / MSPE_RW_gold)
R2_OoS_model_vs_RW_drift_gold = 100 * (1 - MSPE_model_gold / MSPE_RW_drift_gold)


print("Gold Rolling Regression MSPE:", MSPE_model_gold)
print("Gold Random Walk MSPE:", MSPE_RW_gold)
print("Gold RW with drift MSPE:", MSPE_RW_drift_gold)
print("R_OoS vs Random Walk no drift (%):", R2_OoS_model_vs_RW_gold)
print("R_OoS vs RW with drift (%):", R2_OoS_model_vs_RW_drift_gold)
```

```python
# MSPE Lasso
MSPE_lasso_ng      = ((y_true_ng - y_ng_forecast_lasso) ** 2).mean()
MSPE_ols_ng        = ((y_true_ng - y_naturalgas_forecast) ** 2).mean()
MSPE_RW_ng         = ((y_true_ng - rw_pred_ng) ** 2).mean()
MSPE_RW_drift_ng   = ((y_true_ng - rw_drift_pred_ng) ** 2).mean()

# R_OoS
R2_OoS_lasso_ng_RW = 100 * (1 - MSPE_lasso_ng / MSPE_RW_ng)
R2_OoS_lasso_ng_RW_drift = 100 * (1 - MSPE_lasso_ng / MSPE_RW_drift_ng)
R2_OoS_OLS_ng_RW   = 100 * (1 - MSPE_ols_ng / MSPE_RW_ng)
R2_OoS_OLS_ng_RW_drift = 100 * (1 - MSPE_ols_ng / MSPE_RW_drift_ng)

print("Natural Gas  R_OoS Lasso vs RW:", R2_OoS_lasso_ng_RW)
print("Natural Gas  R_OoS Lasso vs RW + drift:", R2_OoS_lasso_ng_RW_drift)
print("Natural Gas  R_OoS OLS vs RW:", R2_OoS_OLS_ng_RW)
print("Natural Gas  R_OoS OLS vs RW + drift:", R2_OoS_OLS_ng_RW_drift)
```

## Ridge Regression

```python
# Setup: datasets and benchmarks
# ============================

commodities = {
    "Gold": {
        "X": X_gold,
        "y": y_gold,
        "rw": rw_pred_gold,
        "rw_drift": rw_drift_pred_gold
    },
    "Corn": {
        "X": X_corn,
        "y": y_corn,
        "rw": rw_pred_corn,
        "rw_drift": rw_drift_pred_corn
    },
    "Natural Gas": {
        "X": X_natural_gas,
        "y": y_natural_gas,
        "rw": rw_pred_ng,
        "rw_drift": rw_drift_pred_ng
    }
}

# Containers
ridge_forecasts = {}
ridge_betas = {}
results = {}

# ============================
# Rolling / Expanding Ridge Regression
# ============================

for name, data in commodities.items():

    X = data["X"]
    y = data["y"]

    n = len(X)
    split = int(0.9 * n)

    y_forecast_ridge = []
    betas_ridge = []

    for i in range(split, n):

        # Expanding window
        X_train = X.iloc[:i]
        y_train = y.iloc[:i]
        X_test  = X.iloc[i:i+1]

        # Standardization (training only)
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled  = scaler.transform(X_test)

        # Ridge with CV
        ridge = RidgeCV(alphas=[0.01, 0.1, 1, 10, 100])
        ridge.fit(X_train_scaled, y_train)

        # One-step-ahead forecast
        y_hat = ridge.predict(X_test_scaled)[0]
        y_forecast_ridge.append(y_hat)

        # Store coefficients
        betas_ridge.append(ridge.coef_)

    # Save forecasts and coefficients
    y_forecast_ridge = pd.Series(
        y_forecast_ridge,
        index=y.iloc[split:].index
    )

    ridge_forecasts[name] = y_forecast_ridge
    ridge_betas[name] = pd.DataFrame(
        betas_ridge,
        columns=X.columns,
        index=X.index[split:]
    )

    # ============================
    # Evaluation
    # ============================

    y_true = y.iloc[split:]
    rw_pred_ridge = data["rw"].loc[y_true.index]
    rw_drift_pred_ridge = data["rw_drift"].loc[y_true.index]

    MSPE_ridge = ((y_true - y_forecast_ridge) ** 2).mean()
    MSPE_RW_ridge = ((y_true - rw_pred_ridge) ** 2).mean()
    MSPE_RW_drift_ridge = ((y_true - rw_drift_pred_ridge) ** 2).mean()

    R2_OoS_RW_ridge = 100 * (1 - MSPE_ridge / MSPE_RW_ridge)
    R2_OoS_RW_drift_ridge = 100 * (1 - MSPE_ridge / MSPE_RW_drift_ridge)

    results[name] = {
        "MSPE_Ridge": MSPE_ridge,
        "MSPE_RW_ridge": MSPE_RW_ridge,
        "MSPE_RW_drift_ridge": MSPE_RW_drift_ridge,
        "R2_OoS_vs_RW_ridge (%)": R2_OoS_RW_ridge,
        "R2_OoS_vs_RW_drift_ridge (%)": R2_OoS_RW_drift_ridge
    }

# ============================
# Results table
# ============================

results_df = pd.DataFrame(results).T
print("\nRidge Regression  Out-of-Sample Performance")
print(results_df)
```
